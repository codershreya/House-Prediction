```python
new = data_frame.groupby("bedrooms")[["price"]].agg({"price":"median"})
new
```

```python

# theta_1 = ((np.mean(new.index)-new.index)*(new["price"].mean()-new["price"])).sum() / ((np.mean(new.index)-new.index)**2).sum()
# theta_0 = new["price"].mean() - (theta_1*np.mean(new.index))
theta_1 = ((np.mean(new.index)-new.index)*(new["price"].mean()-new["price"])).sum() / ((new["price"].mean()-new["price"])**2).sum()
theta_0 = np.mean(new.index) - (theta_1*new["price"].mean())
theta_1, theta_0
```

```python
y_hat = theta_0 + theta_1*new["price"]
y_hat
```

```python
plt.scatter(new, new.index, alpha=0.9, label=f"r = {r:0.2f}")
plt.plot(new['price'], y_hat, color="red", label="Mean Prediction")
plt.xlabel("Price of House's")
plt.ylabel("No.of Bedrooms")
plt.legend(loc=2)
plt.title("Bedroom vs House Price")
plt.show()
```

```python
fig, ((ax1,ax2),(ax3,ax4),(ax5,ax6)) = plt.subplots(nrows=3, ncols=2, figsize=(20, 20), dpi=500)
                                            
seq = zip(
        ["area","bedrooms","bathrooms","parking","stories","mainroad"],
        [ax1   ,   ax2    ,    ax3  ,  ax4     ,    ax5   ,  ax6])
   
for column , ax in seq:
        r = np.corrcoef(data_frame[column], data_frame["price"])[0, 1]
        ax.plot(data_frame[column], data_frame["price"], "bo", label = f"r = {r:0.2f}", mfc='white', alpha=0.5)
        ax.set_title(f"{column} vs Price")
        ax.set_xlabel(column)
        ax.set_ylabel("price")
        ax.legend()
        
plt.show()                       
```
